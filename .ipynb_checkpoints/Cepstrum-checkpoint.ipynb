```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = np.array([1, 3, 4, 2, 1, 9])
fp = np.pad(f, (0, 122))
print(fp)
fhat = np.fft.fft(fp)
c = np.abs(np.fft.fft(np.log(np.abs(fhat))))
```

```python
plt.plot(np.abs(fhat))
plt.show()
```

```python
plt.plot(c)
plt.show()
```

```python
T = 50
omega = np.fft.fftfreq(128, 1)
e = np.exp(-3j*omega*T) + np.exp(3j*omega*T) + np.exp(-2j*omega*T) + np.exp(2j*omega*T) + np.exp(-1j*omega*T) + np.exp(1j*omega*T) + 1
```

```python
#e = np.random.normal(loc=0, scale=1, size=128)
```

```python
ce = np.abs(np.fft.fft(np.log(np.abs(e))))
```

```python
plt.plot(c[:64] + ce[:64])
plt.show()
```
